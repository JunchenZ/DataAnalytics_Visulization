***Step 1***

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score, KFold


def readData(file):    
    df = pd.read_csv(file)
    y = np.zeros(len(df[['m_label']]))
    for i in range(len(df[['m_label']])):
        y[i] = df['m_label'][i]
        
    header = list(df)
    header = header[12:]
    X = np.zeros((len(y), 20, 20))
    for i in range(len(y)):
        for j in range(20):
            for k in range(20):
                X[i][j][k] = df[header[j * 20 + k]][i] / 255.
                
    return X, y
    
X, y = readData('fonts/AGENCY.csv')
```

***Step 2***

```python
def onehotEncode(y):
    data = y
    chars = list(set(data))
    data_size, vocab_size = len(data), len(chars)
    print ('data has %d characters, %d unique.' % (data_size, vocab_size))
    char_to_ix = { ch:i for i,ch in enumerate(chars) }
    ix_to_char = { i:ch for i,ch in enumerate(chars) }
    onehot_encoded = list()
    for value in data:
        letter = [0 for _ in range(len(chars))]
        letter[char_to_ix[value]] = 1
        onehot_encoded.append(letter)
    return onehot_encoded, chars

y, chars = onehotEncode(y)
```

***Step 3***

```python
def makeModel(chars):
    model = Sequential()
    model.add(Conv2D(32, (4, 4), activation='relu', input_shape=(20, 20, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (4, 4)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dropout(0.4))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(len(chars), activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model

model = makeModel(chars)
```

***Step 4***

```python
def getSets(X, y):
    X = np.reshape(X, (-1, 20, 20, 1))
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)
    return X_train, X_test, y_train, y_test

def trainModel(model, X_train, y_train):
    model.fit(np.array(X_train), np.array(y_train), epochs=50, batch_size=32)
    scores = model.evaluate(np.array(X_train), np.array(y_train), batch_size=32)
    print("\ntraining score:")
    print(scores)
    
def testModel(model, X_test, y_test):
    scores = model.evaluate(np.array(X_test), np.array(y_test), batch_size=32)
    print("\ntestining score:")
    print(scores)
    
X_train, X_test, y_train, y_test = getSets(X, y)
trainModel(model, X_train, y_train)
testModel(model, X_test, y_test)
```

```python
def makeBetterModel(chars):
    model = Sequential()
    model.add(Conv2D(32, (4, 4), activation='relu', input_shape=(20, 20, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (4, 4)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dropout(0.25))
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.25))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(len(chars), activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model
    
betterModel = makeBetterModel(chars)
trainModel(betterModel, X_train, y_train)
testModel(betterModel, X_test, y_test)
```

From above we can see that the betterModel actually performs better (about 47%) than the previous one (about 40%)

```python
def newEncode(data, chars):
    data_size, vocab_size = len(data), len(chars)
    print ('data has %d characters, %d unique.' % (data_size, vocab_size))
    char_to_ix = { ch:i for i,ch in enumerate(chars) }
    ix_to_char = { i:ch for i,ch in enumerate(chars) }
    onehot_encoded = list()
    for value in data:
        letter = [0 for _ in range(len(chars))]
        letter[char_to_ix[value]] = 1
        onehot_encoded.append(letter)
    return onehot_encoded

X1, y1 = readData('fonts/AGENCY.csv')
X2, y2 = readData('fonts/English.csv')

y_combine = np.append(y1, y2)
chars = list(set(y_combine))

y1 = newEncode(y1, chars)
betterModel = makeBetterModel(chars)
X1 = np.reshape(X1, (-1, 20, 20, 1))
trainModel(betterModel, X1, y1)

y2 = newEncode(y2, chars)
X2 = np.reshape(X2, (-1, 20, 20, 1))
testModel(betterModel, X2, y2)
```

```python
X1, y1 = readData('fonts/AGENCY.csv')
X2, y2 = readData('fonts/BELL.csv')
X3, y3 = readData('fonts/English.csv')

y_combine = np.append(y1, y2)
y_combine = np.append(y_combine, y3)

chars = list(set(y_combine))
betterModel = makeBetterModel(chars)

X_train = np.append(X1, X2)
y_train = np.append(y1, y2)
y_train = newEncode(y_train, chars)

X_train = np.reshape(X_train, (-1, 20, 20, 1))
trainModel(betterModel, X_train, y_train)

y_test = newEncode(y3, chars)
X_test = np.reshape(X3, (-1, 20, 20, 1))
testModel(betterModel, X_test, y_test)
```

From the above we can see that when training on two differnt sets (one more set than before) and testing on same set as before, the performance gets better from about 19% to about 35%

```python
import matplotlib.pyplot as plt
%matplotlib inline

y_pred = betterModel.predict(X_test, batch_size = 32, verbose = 0, steps=None)
mis = np.zeros(100)
count = 0
for i in range(len(y_pred)):
    if (y_pred[i] != y_test[i]).all():
        mis[count] = i
    count += 1
    if count == 100:
        break
        
plt.figure(figsize= (10, 10))    
for i in mis:
    ii = int(i)
    plt.subplot(10, 10, ii+1)
    plt.imshow(np.reshape(X3[ii],(20, 20)), cmap='Greys',interpolation='nearest')
    plt.axis('off')
plt.show()
```
