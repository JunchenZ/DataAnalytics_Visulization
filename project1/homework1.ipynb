## Question 1

```python
def computStats (data):
    _sum = 0
    for i in data:
        _sum += i
    mean = _sum / len(data)
    print("user function:\nmean:", mean)
    
    _sum2 = 0
    for i in data:
        _sum2 += (i - mean) ** 2
    std = (_sum2 / len(data)) ** (.5)
    print("standard deviation:", std)  
```

```python
from scipy.stats import norm
import numpy as np
mu = 1
sigma = 10
data = norm.rvs(loc = mu, scale = sigma, size = 1000000)
computStats(data)
print("\nnumpy function:\nmean:", np.mean(data))
print("standard deviation:", np.std(data))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(data, bins=100)
plt.show()
```

## Question 2

```python
import pandas as pd
data2 = pd.read_csv("2017-PM2.5.csv")
data2
```

```python
_v4 = data2["V4"]
```

```python
import matplotlib.pyplot as plt
v4 = [.0] * len(_v4)
_v4 = _v4.fillna(0)
for i in range(len(_v4)):
    try:
        v4[i] = float(_v4[i])
    except:
        v4[i] = .0
plt.plot(v4)
```

```python
month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
monthData = [.0] * 12
index = 2
for i in range(12):
    _sum = 0
    count = 0
    for j in range(month[i]):
        for k in range(24):
            if float(v4[index]) > 0:
                _sum += float(v4[index])
                count += 1
            index += 1
    if count != 0:
        monthData[i] = _sum / count
monthData
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.bar(months, monthData)
```

The PM 2.5 level is higher in summer and winter, while lower in spring and fall.

```python
index = 2
hour = [.0] * 24 
hourCount = [0] * 24
for i in range(365):
    for j in range(24):
        if float(v4[index]) > 0:
            hour[j] += float(v4[index])
            hourCount[j] += 1
        index += 1
for i in range(24):
    hour[i] /= hourCount[i]
plt.bar(range(24), hour)
```

PM 2.5 level is higher in the night and lower in the morning and afternoon.

```python
boxData1 = [[0] * 24 * month[i] for i in range(12)]
index = 2
for i in range(12):
    count = 0
    for j in range(month[i]):
        for k in range(24):
            if float(v4[index]) > 0:
                boxData1[i][count] = float(v4[index])
            index += 1
            count += 1
plt.boxplot(boxData1, sym='')
```

In winter and summer, at least half of days have relatively high PM 2.5

```python
boxData2 = [[0] * 365 for i in range(24)]
index = 2
count = 0
for i in range(12):
    for j in range(month[i]):
        for k in range(24):
            if float(v4[index]) > 0:
                boxData2[k][count] = float(v4[index])
            index += 1
        count += 1
plt.boxplot(boxData2, sym='')
```

In the night, the PM 2.5 is relatively high while it is lower in the day time.
