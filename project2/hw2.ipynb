# Part 1:
Null hypothesis: B has the same performance as A

```python
import math
from scipy.stats import norm
import matplotlib.pyplot as plt

def z_value(nB): # z_value calculation with B varies
    n = 1000
    nA = 500
    pA = nA / n
    pB = nB / n
    xA = nA * pA
    xB = nB * pB
    p = (xA + xB) / (nA + nB)
    z = (pA - pB) / math.sqrt(p * (1 - p) * (1 / nA + 1 / nB))
    print("z_value:", z)
    return z

def p_value(z):
    p = norm.cdf(z)
    print("p_value:", p)
    return p
    
z = z_value(550)
p = p_value(z)
```

### Conclusion:
Since the p_value is about 0.053, which is bigger than both 0.01 and 0.05, so we cannot reject the null hypothesis, which means A may have same performance as B. 

Now, if we change the nB to 560 and 570:

```python
z = z_value(560)
p = p_value(z)

z = z_value(570)
p = p_value(z)
```

### Conclusion:
In these two cases, if we choose 0.01 as the significance level, the conclusion is still same as above, which menas we cannot reject the null hypothesis and A and B may have same performance, too.

However, if we choose 0.05 as the significance level, the p_value we have is less than 0.05, so we cannot accept the null hypothesis, which means B may perform better than A do.
